# Simulación Total Ingresos

```python
# Obtenemos el rendimiento simple
Total_ingresos=ingresos.drop(columns=ingresos.columns[1:12])
Ri_ingresos = (Total_ingresos.pct_change(1).iloc[1:])
# Obtenemos su media y desviación estándar de los rendimientos
mu_R_ingresos = Ri_ingresos.mean()[0]
sigma_R_ingresos = Ri_ingresos.std()[0]
Ri_ingresos.plot()
Ri_ingresos.tail()
```

```python
from datetime import date
# Encontrar la fecha de hoy en formato timestamp
today = '2020'
# Obtener fecha de cierre de la opción a valuar
expiry = '2030'
nscen = 5
# Generar rangos de fechas de días hábiles
dates = (2020,2021,2022,2023,2024,2025,2026,2027,2028,2029,2030)
ndays = len(dates)
```

```python
# Simular los rendimientos
# Rendimiento diario 
dt_ingresos = .1
# Z ~ N(0,1) normal estándar (ndays, nscen)
Z_ingresos = np.random.randn(ndays, nscen)

# Simulación normal de los rendimientos
Ri_dt_ingresos = pd.DataFrame(mu_R_ingresos * dt_ingresos + Z_ingresos * sigma_R_ingresos * np.sqrt(dt_ingresos), index=dates)
```

```python
# Simulación del precio
S_0_ingresos = Total_ingresos.iloc[-1,0]
S_T_ingresos = S_0_ingresos*(1+Ri_dt_ingresos).cumprod()
S_T_ingresos

Simulacion_ingreso=pd.DataFrame(index=np.arange(1989,2031),columns=['Total','Total1','Total2','Total3','Total4'])
Simulacion_ingreso['Total']= np.array(pd.concat([Total_ingresos.iloc[:,0],S_T_ingresos.iloc[:,0]]))
Simulacion_ingreso['Total1']= np.array(pd.concat([Total_ingresos.iloc[:,0],S_T_ingresos.iloc[:,1]]))
Simulacion_ingreso['Total2']= np.array(pd.concat([Total_ingresos.iloc[:,0],S_T_ingresos.iloc[:,2]]))
Simulacion_ingreso['Total3']= np.array(pd.concat([Total_ingresos.iloc[:,0],S_T_ingresos.iloc[:,3]]))
Simulacion_ingreso['Total4']= np.array(pd.concat([Total_ingresos.iloc[:,0],S_T_ingresos.iloc[:,4]]))

# Se muestran los precios simulados con los precios descargados
plt.figure(figsize=(18,6))
plt.plot(Simulacion_ingreso.iloc[:,0][0:31],'g',label='Datos originales')
plt.plot(Simulacion_ingreso.iloc[:,0][30:42],'orange',label='Simulación 1')
plt.plot(Simulacion_ingreso.iloc[:,1][30:42],'black',label='Simulación 2')
plt.plot(Simulacion_ingreso.iloc[:,2][30:42],'purple',label='Simulación 3')
plt.plot(Simulacion_ingreso.iloc[:,3][30:42],'brown',label='Simulación 4')
plt.plot(Simulacion_ingreso.iloc[:,4][30:42],'gray',label='Simulación 5')
plt.plot(2019,Total_ingresos.iloc[-1],'ob',label='Ultimo registro de ingreso')
plt.plot(2020,7.5914903948e10,'or',label='Ingreso real 2020 Michoacán')
plt.xlabel('Año')
plt.ylabel('Ingreso a escala')
plt.title('Simulación de precios usando rendimiento simple');
plt.legend();
plt.show()
```

```python
print('Ingreso real Michoacán 2019=',format(int(Total_ingresos.iloc[-1]),'0,.1f'))
print('Ingreso real Michoacán 2020=',format(7.5914903948e10,'0,.1f'))
print('Ingreso simulación 1 2020 Michoacán =',format(int(Simulacion_ingreso.iloc[:,0][31:32]),'0,.1f'))
print('Ingreso simulación 2 2020 Michoacán =',format(int(Simulacion_ingreso.iloc[:,1][31:32]),'0,.1f'))
print('Ingreso simulación 3 2020 Michoacán =',format(int(Simulacion_ingreso.iloc[:,2][31:32]),'0,.1f'))
print('Ingreso simulación 4 2020 Michoacán =',format(int(Simulacion_ingreso.iloc[:,3][31:32]),'0,.1f'))
print('Ingreso simulación 5 2020 Michoacán =',format(int(Simulacion_ingreso.iloc[:,4][31:32]),'0,.1f'))
```

```python
print('Error simulación 1 2020 Michoacán =',np.abs(int(Simulacion_ingreso.iloc[:,0][31:32])/7.5914903948e10-1)*100,'%')
print('Error simulación 2 2020 Michoacán =',np.abs(int(Simulacion_ingreso.iloc[:,1][31:32])/7.5914903948e10-1)*100,'%')
print('Error simulación 3 2020 Michoacán =',np.abs(int(Simulacion_ingreso.iloc[:,2][31:32])/7.5914903948e10-1)*100,'%')
print('Error simulación 4 2020 Michoacán =',np.abs(int(Simulacion_ingreso.iloc[:,3][31:32])/7.5914903948e10-1)*100,'%')
print('Error simulación 5 2020 Michoacán =',np.abs(int(Simulacion_ingreso.iloc[:,4][31:32])/7.5914903948e10-1)*100,'%')
```

```python
# Simulación del precio
S_0_ingresos = Total_ingresos.iloc[-1,0]
S_T_ingresos = S_0_ingresos*(1+Ri_dt_ingresos).cumprod()
S_T_ingresos

Simulacion_ingreso=pd.DataFrame(index=np.arange(1989,2031),columns=['Total','Total1','Total2','Total3','Total4'])
Simulacion_ingreso['Total']= np.array(pd.concat([Total_ingresos.iloc[:,0],S_T_ingresos.iloc[:,0]]))
Simulacion_ingreso['Total1']= np.array(pd.concat([Total_ingresos.iloc[:,0],S_T_ingresos.iloc[:,1]]))
Simulacion_ingreso['Total2']= np.array(pd.concat([Total_ingresos.iloc[:,0],S_T_ingresos.iloc[:,2]]))
Simulacion_ingreso['Total3']= np.array(pd.concat([Total_ingresos.iloc[:,0],S_T_ingresos.iloc[:,3]]))
Simulacion_ingreso['Total4']= np.array(pd.concat([Total_ingresos.iloc[:,0],S_T_ingresos.iloc[:,4]]))

# Se muestran los precios simulados con los precios descargados
plt.figure(figsize=(18,6))
plt.plot(Simulacion_ingreso.iloc[:,0][0:31],'g',label='Datos originales')
plt.plot(Simulacion_ingreso.iloc[:,0][30:42],'orange',label='Simulación 1')
plt.plot(Simulacion_ingreso.iloc[:,1][30:42],'black',label='Simulación 2')
plt.plot(Simulacion_ingreso.iloc[:,2][30:42],'purple',label='Simulación 3')
plt.plot(Simulacion_ingreso.iloc[:,3][30:42],'brown',label='Simulación 4')
plt.plot(Simulacion_ingreso.iloc[:,4][30:42],'gray',label='Simulación 5')
plt.plot(2019,Total_ingresos.iloc[-1],'ob',label='Ultimo registro de ingreso')
plt.plot(2020,1.22400420690e11,'or',label='Ingreso real 2020 Jalisco')
plt.xlabel('Año')
plt.ylabel('Ingreso a escala')
plt.title('Simulación de precios usando rendimiento simple');
plt.legend();
plt.show()
```
